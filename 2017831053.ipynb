```
import numpy as np
import matplotlib.pyplot as plt
import random

print('termtest is over 20 marks')
print('the number of students is 50')

marks = np.random.randint(0,20,50)
student = np.arange(1,51,1)
mean=[np.mean(marks)]
mediun=[np.median(marks)]

print('Mean=',mean)
print('Mediun=',mediun)

meanpl=mean*len(student)
mediunpl=mediun*len(student)

plt.plot(student,marks,marker='o',label='data',linewidth=2)
plt.plot(student,meanpl,'g',label='mean',linestyle='--')
plt.plot(student,mediunpl,'r',label="mediun",linestyle='--')

plt.title('The graphical plot of SWE students TT result')
plt.xlabel('Number of students')
plt.ylabel('Term test marks')

plt.legend(loc='upper right')

plt.show()
```
